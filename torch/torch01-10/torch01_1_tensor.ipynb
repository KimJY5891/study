초기화 되지 않은 텐서

```python
import torch
torch.__version__

x=torch.empty(4,2)
print(x)
# 비어서 초기화 하지 않은값
# 실행할때마다 다른 초기화되지 않은 값들이 나온다.
# 우리가 초기화하지 않아서 기준없는 랜덤한 값이 텐서에 있음
```

무작위로 초기화된 텐서

```python
x = torch.rand(4,2)
# rand - 랜덤의 줄임
print(x)
# 초기화는 되었는데 랜덤을 기준으로 초기화됌
```

데이터 타입(dtype)이 long이고, 0으로 채워진 텐서

```python
x = torch.zeros(4,2)
x_long = torch.zeros(4,2,dtype=torch.long)
# long타입이란 정수 데이터 타입 중 하나로서, 64비트 정수를 표현하는 데 사용됩니다
# 정수 라벨(label) 데이터를 다룰 때 주로 사용되며, 예를 들어 분류 모델의 타겟 라벨 데이터를 표현하는 데 사용됩니다.
# 이는 모델의 출력과 실제 라벨 간의 비교 및 평가에 활용됩니다.
print(f'x:{x}')
print(f'x_long :{x_long}')
```

사용자가 입력한 값으로 텐서 초기화

```python
x = torch.tensor([3,2.3])
print(x)
```

2x4크기, double 타입, 1로 채워진 텐서

```python
x = x.new_ones(2,4,dtype=torch.double)
print(x)
```

x와 같은 크기, float 타입, 무작위로 채워진 텐서


```python
x = torch.randn_like(x,dtype=torch.float)
print(x)
# x = torch.randn_like(작성된변수,dtype=torch.float)
# 작성된 변수 자리에 작성된 변수의 텐서 크기를 가짐
# 위에 기존 x는 (2,4)텐서로 데이터 타입은 float로 해줘
```

텐서의 크기 계산
