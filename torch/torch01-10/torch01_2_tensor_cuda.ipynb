# **데이터 타입**


image.png

```python
ft = torch.FloatTensor([1,2,3])
print(ft)
print(ft.dtype)
# float는 기본이 32비트
```

```python
print(ft.short()) # short는 int 16비트
print(ft.int())
print(ft.long()) # long는 int 16비트
# 캐스팅 : 데이터 타입 변환
```

```python
it = torch.IntTensor([1,2,3])
print(it)
print(it.dtype)
```

```python
print(it.float())
print(it.double())
print(it.half())
```

# CUDA Tensors
*   to 메소드를 사용하여 텐서를 어떤 장치(cpu, gpu)로도 옮길 수 있음


```python
x = torch.randn(1) # 랜덤한 값 하나를 뽑아줘
print(x) # tensor([0.2319]) - 4자리 밖에 안보여줌
print(x.item()) # 0.23194685578346252 - 실제값
print(x.dtype) # torch.float32
```

```python
print(torch.cuda.is_available())
# cuda인데, gpu 사용 불가능하면 cpu | 부득이하게 gpu가 연결 안될 경우 cpu로 연결
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)
y = torch.ones_like(x,device=device)
print(y)
x = x.to(device)
print(x)
z = x + y
print(z) # x,y는 둘 다
print(z.to('cpu'))
print(z.to('cpu',torch.double))
```
