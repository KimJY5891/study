# 텐서의 조작(Manipulations)
인덱싱(Indexing) : NumPy처럼 인덱싱 형태로 사용가능

```python

x = torch.Tensor([[1,2],
                [3,4]])
print(x)
print(x[0,0])
print(x[0,1])
print(x[1,0])
print(x[1,1])

print(x[:,0])
print(x[:,1])

print(x[0,:])
print(x[1,:])
```

view : 텐서의 크기(size)나 모양(shape)을 변경
*   기본적으로 변경 전과 후에 텐서 안의 원소 객수가 유지 되어야함
*   -1로 설정되면 계산을 통해 해당 크기 값을 유추



```python
#view라는 메소드를 사용
#view는 torch에서만 사용 가능하다.
x =torch.randn(4,5)
print(x)
y = x.view(20)
print(y)
z = x.view(5,-1) # -1 : 나머지는 네가 알아서 자동으로 계산해줘
print(z)
```

item : 텐서에 값이 단 하나라도 존재하면 숫자값을 얻을 수 있음

```python
x = torch.randn(1)
print(x)
print(x.dtype)
print(x.item())
print(x.dtype)
```

스칼라값이 하나만 존재해야 item() 사용가능
아니면 에러난다.

```python
x = torch.randn(2)
print(x)
print(x.item())
print(x.dtype)
```

squeeze : 차원 축소(제거)

```python
tensor = torch.rand(1,3,3)
print(tensor)
print(tensor.shape)
```

```python
t = tensor.squeeze()
print(t)
print(t.shape)
```

unsqueeze : 차원 증가(생성)

```python
t = torch.rand(3,3)
print(t)
print(t.shape)
```

```python
tensor = t.unsqueeze(dim=0) # 0번째 차원을 기준으로 늘려줘
print(tensor)
print(tensor.shape)
```

```python
tensor = t.unsqueeze(dim=2) # 2번째 차원을 기준으로 늘려줘
print(tensor)
print(tensor.shape)
```
